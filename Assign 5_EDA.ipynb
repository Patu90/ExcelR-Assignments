```python
#Handle missing values appropriately
data.fillna(data.mean(),inplace = True) #fill null values with mean
```

```python
# Detecting outliers using Z-score
def detect_outliers_zscore(data):
    outliers = {}
    for column in data.select_dtypes(include=[np.number]).columns:
        z_scores = (data[column] - data[column].mean()) / data[column].std()
        outliers[column] = data[np.abs(z_scores) > 3][column]
    return outliers

outliers_zscore = detect_outliers_zscore(data)
print("Outliers detected using Z-score method:\n", outliers_zscore)
```

# 4. Pattern Recognition and Insights:

Identifying correlations between variables is an important step in data analysis as it can reveal relationships between different features of a dataset. These relationships can have significant implications for understanding patterns in the data and for making informed decisions based on that data.

 Calculate the Correlation Matrix
A correlation matrix is a table that shows the correlation coefficients between all pairs of numerical variables in a dataset. The correlation coefficient values range from -1 to 1:

1 indicates a perfect positive correlation.
-1 indicates a perfect negative correlation.
0 indicates no correlation.
